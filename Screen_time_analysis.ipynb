**Screen Time Analysis using Python**

Let's begin the task of analyzing screen time by importing the necessary Python libraries.

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

Let's read in the CSV dataset.

```python
screen_data=pd.read_csv("Screentime-App-Details.csv")
screen_data.head(5)
```

Let's check if there are any null values or not.

```python
screen_data.isnull().sum()
```

Based on the result set, our dataset does not have any null values. Now, let's check its descriptive statistics.

In pandas, the `describe() ` function provides summary statistics of `numeric columns` in a DataFrame. It gives you a quick overview of the distribution of your data. Here's what each metric means:

`count:` Number of non-null observations in the column.

`mean:` Average value of the column.

`std: `Standard deviation, which measures the spread of the values around the mean. A higher standard deviation indicates more dispersion in the data.

`min: `Minimum value in the column.

`25%, 50%, 75%:` These are quartiles. The 25th, 50th (median), and 75th percentiles, respectively. For example, 25% of the data falls below the value shown in the 25th percentile.

`max:` Maximum value in the column.

```python
screen_data.describe()
```

"Let's start by analyzing our dataset, employing a bar chart to show the amount of usage along with the date."

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
plt.figure(figsize=(12,5))
sns.barplot(x='Date', y='Usage', hue='App', palette='husl', data=screen_data)
plt.title('The number of device usage picked by the user in a day for both messaging platforms')
plt.xlabel('Date')
plt.ylabel('Usage Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Based on this analysis, I can conclude that on WhatsApp, the number of devices picked by users per day is higher than on Instagram."

**Let's try to find out the relationship between the number of notifications received by the user in a day and the date.**

```python
import seaborn as sns
import matplotlib.pyplot as plt
custom_palette = ["#1f77b4", "#ff7f0e"]
sns.set_style("ticks")
plt.figure(figsize=(12,5))
sns.barplot(x='Date', y='Notifications', hue='App', palette=custom_palette, data=screen_data)
plt.title('The number of Notifications received by the user in a day for both messaging platforms')
plt.xlabel('Date')
plt.ylabel('Number of Notifications')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Now, let's have a look at the number of times the app was opened in a day.**

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
plt.figure(figsize=(12,5))
sns.barplot(x='Date', y='Times opened', hue='App', palette='husl', data=screen_data)
plt.title('The number of device usage picked by the user in a day for both messaging platforms')
plt.xlabel('Date')
plt.ylabel('Number of times app opened')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

This bar illustration demonstrates the relationship between the number of times the app was opened by the user and the date. Based on this result, we can say that the number of times the app was opened on WhatsApp is more than on Instagram.

**Let's plot a scatter plot to display the relationship between the number of notifications received by the user and their usage.**

```python
import seaborn as sns
scatter_kws = {"s": 100}
sns.regplot(data=screen_data, x='Notifications', y='Usage',scatter_kws=scatter_kws, marker='o', color='Black', line_kws={"color": "red"})

plt.xlabel('Notifications')
plt.ylabel('Amount of Usage')
plt.title('Scatter Plot: Notifications vs Amount of Usage with Trendline')
plt.show()
```

**As the number of notifications received by users increases, the number of device usage per day also increases.**
